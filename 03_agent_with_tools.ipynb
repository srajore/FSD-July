```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama
from langchain_community.tools.tavily_search import TavilySearchResults
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = ChatOllama(model="gpt-oss:120b-cloud")
```

```python
tavily_tool = TavilySearchResults(max_results=3)
```

```python
my_agent = create_agent(
    model= llm,
    tools=[tavily_tool],
    system_prompt="you are helpful assistant"
)
```

```python
my_agent
```

```python
response = my_agent.invoke({"messages":"who won yesterdays(14th July 2026) India vs England Cricket ODI match? "})
```

```python
response = my_agent.invoke({"messages":"what are the latest top 3 news in India? "})
```

```python
response
```

```python
from langchain_tavily import TavilySearch
```

```python
# Initialize Tavily Search Tool
tavily_search_tool = TavilySearch(
    max_results=5,
    topic="general",
)

agent = create_agent(model=llm, tools=[tavily_search_tool])

user_input = "What nation hosted the Euro 2024? Include only wikipedia sources."
```

```python
response = agent.invoke({"messages":"what are the latest top 3 news in India?"})
```

```python
response
```

```python
from langchain_community.tools.google_serper import GoogleSerperResults

search_tool = GoogleSerperResults()

agent = create_agent(
    model=llm,
    tools=[search_tool],
)

agent.invoke({"messages":"what is the todays stock prize of Zensar Technologies at BSE?"})
```
