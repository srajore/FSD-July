```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
```

```python
llm = init_chat_model(
    model= os.getenv("MODEL_NAME"),
    model_provider=os.getenv("MODEL_PROVIDER")
)
```

```python
response = llm.invoke("What is the capital of France?")
```
